```python
def print_useful_rows_info(df):
    print('Количество полностью заполенных объектов из всей выборки:', len(df.dropna()))
    print('Процент полностью заполненных объектов из всей выборки:', round(len(df.dropna()) / len(df) * 100, 2))
```

```python
print_useful_rows_info(df_clean)
```

```python
df_clean = df_clean.drop(columns=['type', 'paint_color', 'drive', 'condition', 'VIN', 'cylinders', 'size'])
print_useful_rows_info(df_clean)
```
