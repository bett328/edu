```python
# Генератор бинарного числа
def gen_bin(M, prefix=""):
    '''Генератор бинарных чисел длины M.'''
    if M == 0:
        print(prefix)
        return
    
    gen_bin(M-1, prefix+"0")
    gen_bin(M-1, prefix+"1")
    
gen_bin(3)

# Генератор всех чисел
def generate_number(N:int, M:int, prefix=None):
    '''Генератор всех чисел (с лидирующими незначащими нулями)
    в N-ричной системе счисления (N <= 10) длины M.'''
    
    prefix = prefix or []
    if M == 0:
        print(*prefix)
        return
    for digit in range(N):
        prefix.append(digit)
        generate_number(N, M-1, prefix)
        prefix.pop()
            
generate_number(3, 2)

# Линейный поиск
def find(number, A):
    '''Ищет number в A, возвращает True/False'''
    for x in A:
        if number == x:
            return True
    return False

# Генератор перестановок для множества N
def generate_permitations(N:int, M:int=-1, prefix=None):
    ''' Генератор всех перестановок N чисел в M позициях (СС), 
        начиная с префикса prefix.
    '''
    M = N if M == -1 else M # по умолчанию N чисел в N позициях
    
    prefix = prefix or []
    
    if M == 0:
        print(*prefix)
        return
    
    for number in range(1, N+1):
        if find(number, prefix):
            continue
            
        prefix.append(number)
        generate_permitations(N, M-1, prefix)
        prefix.pop()
        
generate_permitations(3)
```
