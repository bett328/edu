```python
# Линейный поиск в списке: 
def linear_array_search(A:list, N:int, x:int):
    ''' 
    Находит число в списке 
    А - список, N - размер списка, x - искомое число;
    Возвращает индекс найденого числа в списке;
    При отсутствии вовзращает -1 ;
    '''
    for k in range(N):
        if A[k] == x:
            return k
    return -1
            

# Тесты  для линейного поиска по списку
def linear_array_search_test():
    A1 = [1, 2, 3, 4, 5]
    result1 = linear_array_search(A1, 5, 8)
    if result1 == -1:
        print("Test1 passed")
    else:
        print("Test1 failed")
        
    A2 = [-1, -2, -3, -4, -5]
    result2 = linear_array_search(A2, 5, -3)
    if result2 > 0:
        print("Test2 passed")
    else:
        print("Test2 failed")
                                  
linear_array_search_test()
```

```python
# Бинарный поиск по левой/правой границе
def binary_search_left_bound(A, key):
    '''Поиск левой границы от искомого числа'''
    A = sorted(A)
    left = -1
    right = len(A)
    while right - left > 1:
        middle = (left + right) // 2
        if A[middle] < key:
            left = middle
        else:
            right = middle
    return left

def binary_search_right_bound(A, key):
    '''Поиск левой границы от искомого числа'''
    A = sorted(A)
    left = -1
    right = len(A)
    while right - left > 1:
        middle = (left + right) // 2
        if A[middle] <= key:
            left = middle
        else:
            right = middle
    return right
```

```python
# Merging sort - Алгорит сортировки слиянием. 
def merge_sort(A, B):
    ''' Сортировка слиянием. Даны два списка file1 и file2. 
        Cписок result должен сожержать отсортированное объединение file1 и file2.
    '''
    result = []
    i, j = 0
    while i < len(A) and j < len(B):
        if A[i] < B[j]:
            result.append(A[i])
            i += 1
        else:
            result.append(B[j])
            j += 1
    # приклеиваем остаток
    result += A[i:] + B[j:]
```

```python
# Линейный поиск в массиве
def find(number, A):
    '''Ищет number в A, возвращает True/False'''
    for x in A:
        if number == x:
            return True
    return False
```
